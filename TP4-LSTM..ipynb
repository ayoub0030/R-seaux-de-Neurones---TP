Objectifs pédagogiques
 Préparer les données pour le Deep Learning (fenêtrage, normalisation).
 Construire et entraîner un RNN simple, puis un LSTM.
 Comparer leurs performances sur un problème réel de prévision.
L’objectif de ce TP est de construire un modèle de Deep Learning capable de prédire la
température minimale du jour suivant à partir d’historiques de températures quotidiennes.
Dataset : Daily Minimum Temperatures in Melbourne (1981–1990)
Questions
 Pourquoi ce dataset est-il un problème de séries temporelles et non de classification ?
 Quels défis un réseau doit-il gérer pour prévoir une température future ?
 Pourquoi normaliser les données ?
 Pourquoi faut-il construire des fenêtres temporelles ?
 Quel impact du window_size sur la performance ?
 Construction du modèle RNN
 Construction du modèle LSTM
 Tracer la prédiction vs la vérité terrain sur le jeu de test.

1-Chaque donnée dépend du temps (température du jour J dépend de J-1, J-2…).

Les valeurs sont continues (des nombres réels) → ce n’est pas une étiquette ou une catégorie.

2 -La dépendance temporelle .  Les patterns non linéaires

3- Les réseaux (RNN/LSTM) apprennent mieux si toutes les valeurs sont dans une même échelle.

4-Car les réseaux ne peuvent pas deviner seuls combien d’historique utiliser.

5-un window size de 30 jours est très bon pour ces données meeteo : meilleur équilibre

capte tendance + saison courte
